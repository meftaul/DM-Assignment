# Loading the data

````python
import pandas as pd
data = pd.read_csv('Assignment2-Modified.csv.csv')
data.head()
````

Prior to loading data in this notebook we have done some preprocessing in the excel sheet. 

```How much time do you spend daily on social media (e.g., Facebook)?``` converted to numeric values in hour.

```How much time do you spend each day on an average in the online classes?``` this column converted to numeric values - average hour. (e.g -> 3-5hour = 4 hour)

**We have 50 rows and 35 columns in the datasets. Column names are too long to handle so first of we renamed the columns for easier management.** 

````python
column_names = data.columns.to_list()
new_column_names = ['timestamp', 
 'gender', 
 'age',
 'permanent_address',
 'monthly_internet_cost',
 'monthly_call_cost',
 'time_on_fb',
 'difficult_to_manage_expense',
 'household_monthly_income',
 'regular_in_online_class',
 'physical_problem_due_to_online_class',
 'physical_problem_description',
 'device_for_online_class',
 'avg_time_spent_online_class',
 'remote_learning_effectiveness',
 'help_from_university',
 'time_management_for_remote_learning',
 'teacher_helpfulness',
 'class_material_helpful',
 'want_video_recording',
 'enjoy_remote_learning',
 'prefer_structured_course',
 'has_setup_for_interation',
 'teacher_student_interaction',
 'quality_of_discussion',
 'lecture_material_valuable',
 'lecture_material_support_goal',
 'video_recording_not_important',
 'oncampus_online_class_same',
 'learn_more_online',
 'prefer_online_course',
 'receive_online_homework',
 'receive_online_assignment',
 'want_to_participate_online_class',
 'want_to_participate_online_quiz']

column_definition = {}
for key in column_names:
    for value in new_column_names:
        column_definition[key] = value
        new_column_names.remove(value)
        break
column_definition
new_data = data.rename(columns=column_definition)
````
